```python
import pandas as pd

# Load the CSV files into DataFrames
df_main = pd.read_csv('Mens_Tennis_Grand_Slam_Winner.csv')
df_countries = pd.read_csv('countries.csv')
df_locations = pd.read_csv('world_coordinates.csv')

# Merge DataFrames with different column names
# Suppose 'main_id' is the column in df_main
# 'country_id' is the column in df_countries
# and 'location_id' is the column in df_locations

# Merge df_main with df_countries
df_merged = pd.merge(df_main, df_countries, left_on='WINNER_NATIONALITY', right_on='nationality', how='left')

# Merge the result with df_locations
df_merged = pd.merge(df_merged, df_locations, left_on='alpha_2_code', right_on='Code', how='left')

# Save the updated DataFrame to a new CSV file
df_merged.to_csv('updated_main_file.csv', index=False)

# Convert the DataFrame to JSON format and save it
df_merged.to_json('updated_main_file.json', orient='records', lines=False)
```
